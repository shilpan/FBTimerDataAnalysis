```python
import json
import datetime
json_data = open('data/user_document.json')
fb_timer_data = json.load(json_data)
```

```python
total_user_time = {}
overall_time = 0
for user in fb_timer_data:
    overall_user_time = 0
    for session in user['userSessions']:
        total_session_time = 0
        session_start_time = 0
        is_first_session = True
        is_session_empty = True
        for pageSession in session['pageSessions']:
            if 'sessionEnd' in pageSession:
                total_pagesession_time = pageSession['sessionEnd'] - pageSession['sessionStart']
                if (total_pagesession_time > 6000000):
                    total_pagesession_time = 6000000
                total_session_time += total_pagesession_time
                total_user_time.setdefault(user['personID'], {}).setdefault('pagesessions', []).append(total_pagesession_time/60000)
                total_user_time.setdefault(user['personID'], {}).setdefault('pagesessions_date', []).append(datetime.datetime.fromtimestamp(int(pageSession['sessionStart']/1000.0)))
                if (is_first_session):
                    session_start_time = pageSession['sessionStart']
                    is_first_session = False
                    is_session_empty = False
        
        if (is_session_empty != True):
            total_user_time.setdefault(user['personID'], {}).setdefault('sessions', []).append(total_session_time/60000)
            total_user_time.setdefault(user['personID'], {}).setdefault('session_date', []).append(datetime.datetime.fromtimestamp(int(session_start_time/1000.0)))
            overall_user_time += total_session_time
    total_user_time[user['personID']]['overallTime'] = overall_user_time
    overall_time += overall_user_time
```
